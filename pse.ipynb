# 周期的二乗指数型共分散関数に関する考察

簡単のため，ここでは $\Omega=[a,b]=[0,1]$ （従って $\rho=1$）とする．

## 固有値のプロット

固有値 $\kappa_{j}=\exp\left(-(2\pi l)^{-2}\right)I_{\lfloor(j+1)/2\rfloor}\left((2\pi l)^{-2}\right)$ は，$\sum_{j=0}^\infty\kappa_j=1$ なので，非負整数上の確率分布と見なすこともできる．いくつかのパラメータでそれらの値をプロットしてみる．

```matlab
a=0;b=1;
param.domain=[a,b];
param.v = 1;
L = [1/4,1/8,1/16];
q = 11;
figure; hold on
for i=1:length(L)
   param.l = L(i);
   evs = pse_eigenvalues(q,param);
   plot(0:q-1,evs,'o')
end
hold off
xlabel('$j$','Interpreter','latex')
ylabel('$\kappa_j$','Interpreter','latex')
```

$j=2s-1$ とすると固有関数 $\phi_j(x)=\sqrt{2}\cos(2\pi sx)$ なので，その周波数は $s$．また，$j=2s$ とすると $\phi_j(x) = \sqrt{2}\sin(2\pi sx)$ であり，周波数は $s$．従って，
$$
\begin{align*}
\text{平均周波数} &= \sum_{s=1}^\infty s(\kappa_{2s-1}+\kappa_{2s}) \\
&= \sum_{s=1}^\infty 2s\kappa_{2s}.
\end{align*}
$$
である．この平均周波数と $l$ の逆数関係をプロットしてみると，以下のように期待される通りの比例関係がよく成り立っている．$1/l$ が小い部分，つまり，$l$ が大きいところで若干ずれるが，それは，$l$ が領域幅に近いためだろう．

```matlab
IL = linspace(1,100,51); % l の逆数
mf = zeros(size(IL));
S = 100;
for i=1:length(IL)
    param.l = 1/IL(i);
    evs = pse_eigenvalues(2*S+1,param);
    mf(i) = 2*dot(1:S,evs(2:2:end));
end
plot(IL,mf)
xlabel('$1/l$','Interpreter','latex')
ylabel('平均周波数')
```
